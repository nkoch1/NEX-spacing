# NEX Spacing
## Splits time-based data into groups based on a time interval

This Jupyter Notebook imports a python script that must be in the same directory as the Notebook in order for it to work.

```python
from timespacing import *
```

Load the datafile.

```python
[data, filename] = load_data()
```

Set the delta interval.

```python
delta = user_time_delta()
```

Group the data by time.

```python
adjdata = adjdata(data, delta)
```

Save adjusted data with avg, std, and cov added to new CSV file.

```python
save_data(adjdata, filename)
```
